```python
def fun(*args):
    if len(args)==0:
        return 0
    if len(args)==1:
        return args[0]
    return args[0]+fun(*args[1:])
print(fun(10,20))
```

```python
def make_greeting(name,prefix='Hello',formatter=lambda x:x):
    greeting=f"{prefix},{name}"
    return formatter(greeting)
print(make_greeting("Neha"))
print(make_greeting("Neha",formatter=str.upper))
```

```python
double=lambda x:x*2
triple=lambda x:x*3
quadruple=lambda x:x*4

funcs=[double ,triple ,quadruple]
def apply_all(funcs,value):
    for func in funcs:
        value=func(value)
    return value
print(apply_all(funcs,34))
```
