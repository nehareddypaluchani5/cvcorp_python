```python
double = lambda x: x * 2
triple = lambda x: x * 3
quadruple = lambda x: x * 4
funcs = [double, triple, quadruple]
def apply_all(func, value):
    for func in funcs:
        value = func(value)
    return value
print(apply_all(funcs, 123935))
```
