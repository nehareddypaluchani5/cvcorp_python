```python
def Dec (func):
    print("Function Called")
    def Inner():
        print("Starting function")
        func()
        print("Ending function")
    return Inner
@Dec
def greet():
    print("Hello Bro")
greet()
print(f"greet : {greet.__name__}")
```

```python
def valid(func):
    @functools.wraps(func)
    def inner(a,b):
        if isinstance(a,str):
            a=str(a)
        if not isinstance(b,str):
            b=str(b)
        return func(a,b)
    return inner


@valid
def fun(a:str,b:str)->str:
    """This is a Docstring, This Function is for String Concatenation"""
    return a+b
print(fun.__name__)
print(fun("12","13"))
print(fun.__annotations__)
print(fun.__doc__)
# print(print.__doc__)
```

```python
#LEVEL  MEDIUM FIRST  QUESTION
def Validate_positive(func):
    def inner(*args):
        for i in args:
            if i<0:
                print("error message")
                return
        return func(*args)
    return inner
@Validate_positive
def multiply(a,b):
    return a*b
print(multiply(1,2))
print(multiply(-12,23))
```
